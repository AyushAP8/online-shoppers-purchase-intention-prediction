*We can take 0.1 as threshold for the deciding the correlation between the features and the target variable*

```python
def coeff(x,y):
  cf=np.corrcoef(x,y)
  return(cf[0][1])
```

```python
corrd=data.drop(['Month','VisitorType'],axis=1)
```

```python
for i in corrd.columns:
  #print(i)
  #break
  k=coeff(corrd[i],data['Revenue'])
  h=0.1
  if(np.absolute(k) < h):
    corrd.drop(i,axis=1)
```

```python
corrd.columns
```

```python
sns.pairplot(corrd,hue='Revenue')
```

***From the above heatmap and pairplots we can conclude that Page values is the feature that is most correlated with the target variable Revenue.Remaining all the variables are not highly correlated but cannot be counted as negligible***

**So we have to consider all the features that have correlation higher than threshold**

*We have to perform one hot encoding for the categorical variables like Month and Visitor Type*

```python
x=corrd.drop(['Revenue'],axis=1)
y=corrd['Revenue']
```

```python
x= pd.concat([x,pd.get_dummies(data['Month'], prefix='Month')], axis=1)
x= pd.concat([x,pd.get_dummies(data['VisitorType'], prefix='VisitorType')], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.25, random_state=3)
```

```python
data.dtypes
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_predlr=lr.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_predlr),'%')
print(confusion_matrix(y_test,y_predlr))
print(classification_report(y_test,y_predlr))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
y_predrfc=rfc.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_predrfc),'%')
print(confusion_matrix(y_test,y_predrfc))
print(classification_report(y_test,y_predrfc))
```

```python
from sklearn.neural_network import MLPClassifier
mlp=MLPClassifier(activation='relu',solver='adam',hidden_layer_sizes=(50,))
mlp.fit(x_train,y_train)
```

```python
y_predmlp=mlp.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_predmlp),'%')
print(confusion_matrix(y_test,y_predmlp))
print(classification_report(y_test,y_predmlp))
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=3)
```

```python
x_trainos,y_trainos=sm.fit_resample(x_train,y_train)
```

```python
lr.fit(x_trainos,y_trainos)
```

```python
y_predlr=lr.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_predlr),'%')
print(confusion_matrix(y_test,y_predlr))
print(classification_report(y_test,y_predlr))
```

```python
rfc.fit(x_trainos,y_trainos)
```

```python
y_predrfc=rfc.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_predrfc),'%')
print(confusion_matrix(y_test,y_predrfc))
print(classification_report(y_test,y_predrfc))
```

```python
mlp.fit(x_trainos,y_trainos)
```

```python
y_predmlp=mlp.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_predmlp),'%')
print(confusion_matrix(y_test,y_predmlp))
print(classification_report(y_test,y_predmlp))
```

*Here We can observe that after oversampling there is lots of improvement in the F1 score*

*Lets use Grid SearchCV for hyper parameter tuning of Random Forest model and Neural network*

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'n_estimators':[50,100],
        'max_leaf_nodes':[x for x in range(0,10)],
        'max_depth':[x for x in range(0,20)],
        'criterion':['gini','entropy']}
```

```python
gridsearch=GridSearchCV(rfc,params)
```

```python
gridsearch.fit(x_trainos,y_trainos)
```

```python
gridsearch.best_params_
```

```python
rfc_best=gridsearch.best_estimator_
```

```python
rfc_best.fit(x_trainos,y_trainos)
```

```python
y_prrb=rfc_best.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_prrb),'%')
print(confusion_matrix(y_test,y_prrb))
print(classification_report(y_test,y_prrb))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomsearch=RandomizedSearchCV(rfc,params,cv=3)
```

```python
randomsearch.fit(x_trainos,y_trainos)
```

```python
randomsearch.best_params_
```

```python
rfc_randbest=randomsearch.best_estimator_
```

```python
rfc_randbest.fit(x_trainos,y_trainos)
```

```python
y_pr_rb=rfc_randbest.predict(x_test)
print('Test set Accuracy is',100*accuracy_score(y_test,y_pr_rb),'%')
print(confusion_matrix(y_test,y_pr_rb))
print(classification_report(y_test,y_pr_rb))
```
